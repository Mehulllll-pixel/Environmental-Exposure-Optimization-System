```python
import requests, json, os, pandas as pd
from tqdm.notebook import tqdm

SHEET_CSV = "https://docs.google.com/spreadsheets/d/1bujiO2NgtHlgqPlNvYAQf5_7ZcXARlIfNX5HNb9f8cI/export?format=csv&gid=1786138861"
BASE      = "https://storage.googleapis.com/upload_goai"

os.makedirs('/content/audio', exist_ok=True)
os.makedirs('/content/transcriptions', exist_ok=True)
os.makedirs('/content/clips', exist_ok=True)

df = pd.read_csv(SHEET_CSV)

def extract_session_id(url):
    return url.strip().split('/')[-2]

df['session_id']    = df['transcription_url_gcp'].apply(extract_session_id)
df['trans_url_fixed'] = df.apply(lambda r: f"{BASE}/{r['session_id']}/{r['recording_id']}_transcription.json", axis=1)
df['audio_url_fixed'] = df.apply(lambda r: f"{BASE}/{r['session_id']}/{r['recording_id']}_audio.wav", axis=1)

print(f"Found {len(df)} recordings. Downloading...")

# Download transcriptions
failed = []
for _, row in tqdm(df.iterrows(), total=len(df), desc="Transcriptions"):
    out = f"/content/transcriptions/{row['recording_id']}_transcription.json"
    if os.path.exists(out): continue
    try:
        r = requests.get(row['trans_url_fixed'], timeout=30)
        r.raise_for_status()
        with open(out, 'w', encoding='utf-8') as f:
            json.dump(r.json(), f, ensure_ascii=False)
    except Exception as e:
        failed.append(row['recording_id'])

print(f"Transcriptions: {len(df)-len(failed)}/{len(df)}")

# Download audio
failed_audio = []
for _, row in tqdm(df.iterrows(), total=len(df), desc="Audio"):
    out = f"/content/audio/{row['recording_id']}_audio.wav"
    if os.path.exists(out): continue
    try:
        r = requests.get(row['audio_url_fixed'], timeout=120, stream=True)
        r.raise_for_status()
        with open(out, 'wb') as f:
            for chunk in r.iter_content(8192): f.write(chunk)
    except Exception as e:
        failed_audio.append(row['recording_id'])

print(f"Audio: {len(df)-len(failed_audio)}/{len(df)}")
```

```python
import json, librosa, soundfile as sf, numpy as np
from tqdm.notebook import tqdm

TARGET_SR = 16000
MIN_DUR, MAX_DUR, MIN_CHARS = 1.0, 30.0, 2

records, skipped = [], 0

for trans_file in tqdm(os.listdir('/content/transcriptions')):
    rec_id     = trans_file.replace('_transcription.json', '')
    audio_path = f'/content/audio/{rec_id}_audio.wav'
    trans_path = f'/content/transcriptions/{trans_file}'
    if not os.path.exists(audio_path): continue

    with open(trans_path, encoding='utf-8') as f:
        segments = json.load(f)

    try:
        audio, sr = librosa.load(audio_path, sr=TARGET_SR, mono=True)
    except:
        continue

    for i, seg in enumerate(segments):
        start, end, text = seg['start'], seg['end'], seg['text'].strip()
        dur = end - start
        if dur < MIN_DUR or dur > MAX_DUR or len(text) < MIN_CHARS:
            skipped += 1
            continue
        clip = audio[int(start*TARGET_SR):int(end*TARGET_SR)]
        clip_name = f"{rec_id}_seg{i:04d}.wav"
        sf.write(f'/content/clips/{clip_name}', clip, TARGET_SR)
        records.append({'recording_id': rec_id, 'clip_file': clip_name,
                        'duration': round(dur,3), 'text': text})

df_segs = pd.DataFrame(records)
print(f"Segments: {len(df_segs)} kept, {skipped} skipped")
print(f"Total hours: {df_segs['duration'].sum()/3600:.2f}")

# Train/val split
rec_ids = df_segs['recording_id'].unique().tolist()
np.random.seed(42); np.random.shuffle(rec_ids)
split = int(0.85 * len(rec_ids))
train_ids = set(rec_ids[:split])
df_segs['split'] = df_segs['recording_id'].apply(lambda x: 'train' if x in train_ids else 'val')

train_df = df_segs[df_segs['split']=='train']
val_df   = df_segs[df_segs['split']=='val']
train_df.to_csv('/content/train.csv', index=False)
val_df.to_csv('/content/val.csv',   index=False)
print(f"Train: {len(train_df)} | Val: {len(val_df)}")
```

```python
import torch
from transformers import WhisperProcessor, WhisperForConditionalGeneration
from jiwer import wer
from tqdm.notebook import tqdm

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Device: {DEVICE}")

processor = WhisperProcessor.from_pretrained("openai/whisper-small")
model     = WhisperForConditionalGeneration.from_pretrained("openai/whisper-small").to(DEVICE)
model.eval()

refs, hyps = [], []
for _, row in tqdm(val_df.iterrows(), total=len(val_df)):
    audio, _ = librosa.load(f"/content/clips/{row['clip_file']}", sr=16000)
    inputs   = processor(audio, sampling_rate=16000, return_tensors="pt").input_features.to(DEVICE)
    with torch.no_grad():
        ids = model.generate(inputs, language="hi", task="transcribe", max_new_tokens=225)
    hyp = processor.batch_decode(ids, skip_special_tokens=True)[0].strip()
    refs.append(row['text'].strip())
    hyps.append(hyp)

baseline_wer = wer(refs, hyps)
print(f"\n{'='*40}")
print(f"BASELINE WER: {baseline_wer:.4f} ({baseline_wer*100:.2f}%)")
print(f"{'='*40}")
```

```python
from dataclasses import dataclass
from typing import Any, Dict, List
from torch.utils.data import Dataset as TorchDataset
from transformers import Seq2SeqTrainingArguments, Seq2SeqTrainer
import evaluate

# Reload fresh model for fine-tuning
model_ft = WhisperForConditionalGeneration.from_pretrained("openai/whisper-small")
model_ft.generation_config.language = "hindi"
model_ft.generation_config.task = "transcribe"
model_ft.generation_config.forced_decoder_ids = None

class HindiASRDataset(TorchDataset):
    def __init__(self, df):
        self.df = df.reset_index(drop=True)
    def __len__(self): return len(self.df)
    def __getitem__(self, idx):
        row   = self.df.iloc[idx]
        audio, _ = librosa.load(f"/content/clips/{row['clip_file']}", sr=16000)
        feats = processor(audio, sampling_rate=16000, return_tensors="pt").input_features[0]
        labs  = processor.tokenizer(row['text'], return_tensors="pt").input_ids[0]
        return {"input_features": feats, "labels": labs}

@dataclass
class DataCollator:
    processor: Any
    decoder_start_token_id: int
    def __call__(self, features):
        inp = [{"input_features": f["input_features"]} for f in features]
        batch = self.processor.feature_extractor.pad(inp, return_tensors="pt")
        lbl   = [{"input_ids": f["labels"]} for f in features]
        lbl_b = self.processor.tokenizer.pad(lbl, return_tensors="pt")
        labels = lbl_b["input_ids"].masked_fill(lbl_b.attention_mask.ne(1), -100)
        if (labels[:, 0] == self.decoder_start_token_id).all(): labels = labels[:, 1:]
        batch["labels"] = labels
        return batch

metric = evaluate.load("wer")
def compute_metrics(pred):
    pred_ids  = pred.predictions
    label_ids = pred.label_ids
    label_ids[label_ids == -100] = processor.tokenizer.pad_token_id
    pred_str  = processor.batch_decode(pred_ids,  skip_special_tokens=True)
    label_str = processor.batch_decode(label_ids, skip_special_tokens=True)
    return {"wer": metric.compute(predictions=pred_str, references=label_str)}

DRIVE_MODEL = "/content/drive/MyDrive/josh_talks_asr/models/whisper-hindi"

args = Seq2SeqTrainingArguments(
    output_dir                  = DRIVE_MODEL,
    per_device_train_batch_size = 16,   # T4 can handle larger batches
    gradient_accumulation_steps = 2,
    learning_rate               = 1e-5,
    warmup_steps                = 200,
    max_steps                   = 2000,
    gradient_checkpointing      = True,
    fp16                        = True,
    eval_strategy               = "steps",
    eval_steps                  = 200,
    save_steps                  = 200,
    logging_steps               = 25,
    load_best_model_at_end      = True,
    metric_for_best_model       = "wer",
    greater_is_better           = False,
    predict_with_generate       = True,
    generation_max_length       = 225,
    report_to                   = ["none"],
    save_total_limit            = 2,
)

trainer = Seq2SeqTrainer(
    model            = model_ft,
    args             = args,
    train_dataset    = HindiASRDataset(train_df),
    eval_dataset     = HindiASRDataset(val_df),
    data_collator    = DataCollator(processor, model_ft.config.decoder_start_token_id),
    compute_metrics  = compute_metrics,
    processing_class = processor.feature_extractor,
)

print("Starting fine-tuning on T4 GPU...")
trainer.train()
trainer.save_model(DRIVE_MODEL)
processor.save_pretrained(DRIVE_MODEL)
print("Done! Model saved to Google Drive.")
```

```python
from tqdm.notebook import tqdm
import librosa, torch

print("Caching all audio into RAM (~8 mins)...")
train_feats, train_labs = [], []
val_feats, val_labs = [], []

for _, row in tqdm(train_df.iterrows(), total=len(train_df), desc="Train"):
    audio, _ = librosa.load(f"/content/clips/{row['clip_file']}", sr=16000)
    train_feats.append(processor(audio, sampling_rate=16000, return_tensors="pt").input_features[0])
    train_labs.append(processor.tokenizer(row['text'], return_tensors="pt").input_ids[0])

for _, row in tqdm(val_df.iterrows(), total=len(val_df), desc="Val"):
    audio, _ = librosa.load(f"/content/clips/{row['clip_file']}", sr=16000)
    val_feats.append(processor(audio, sampling_rate=16000, return_tensors="pt").input_features[0])
    val_labs.append(processor.tokenizer(row['text'], return_tensors="pt").input_ids[0])

class CachedDataset(torch.utils.data.Dataset):
    def __init__(self, f, l): self.f, self.l = f, l
    def __len__(self): return len(self.f)
    def __getitem__(self, i): return {"input_features": self.f[i], "labels": self.l[i]}

train_dataset = CachedDataset(train_feats, train_labs)
val_dataset   = CachedDataset(val_feats,   val_labs)
print(f"Done! Train: {len(train_dataset)} | Val: {len(val_dataset)}")
```

```python
# Stop current training first (Runtime → Interrupt)
# Then run this to pre-cache all features into RAM

import librosa
import torch
import numpy as np
from tqdm.notebook import tqdm

print("Pre-caching all audio features into RAM...")

def cache_split(df):
    features, labels = [], []
    for _, row in tqdm(df.iterrows(), total=len(df)):
        audio, _ = librosa.load(f"/content/clips/{row['clip_file']}", sr=16000)
        feat = processor(audio, sampling_rate=16000, return_tensors="pt").input_features[0]
        lab  = processor.tokenizer(row['text'], return_tensors="pt").input_ids[0]
        features.append(feat)
        labels.append(lab)
    return features, labels

train_features, train_labels = cache_split(train_df)
print("Train cached!")
val_features, val_labels = cache_split(val_df)
print("Val cached!")

# Now redefine dataset using RAM cache
class CachedDataset(torch.utils.data.Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels   = labels
    def __len__(self): return len(self.features)
    def __getitem__(self, idx):
        return {"input_features": self.features[idx],
                "labels":         self.labels[idx]}

train_dataset_cached = CachedDataset(train_features, train_labels)
val_dataset_cached   = CachedDataset(val_features,   val_labels)
print(f"Ready! Train: {len(train_dataset_cached)} | Val: {len(val_dataset_cached)}")
```
